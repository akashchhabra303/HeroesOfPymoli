* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_lst=[0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,1000]
lst_name = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase = purchase_data.loc[:,["Price","Item ID","Age"]]
purchase = purchase.drop_duplicates()


purchase["age_ranges"] = pd.cut(purchase["Age"], age_lst, labels= lst_name)


purchase_count = purchase.groupby(["age_ranges"]).count()["Item ID"]
purchase_count

count_purchase_analysis = purchase["age_ranges"].value_counts()
count_purchase_analysis

total_purchase = purchase.groupby(["age_ranges"]).sum()["Price"]
total_purchase

avg_purchase = total_purchase / count_purchase_analysis
avg_purchase

purchase_pp = total_purchase/count_dem
purchase_pp

Purchase_df = pd.DataFrame ({"Purchase Count" : purchase_count, "Average Purchase Price" : avg_purchase,
                                  "Total Purchase Value" : total_purchase, "Purchase Per Person" : purchase_pp} )






Purchase_df['Average Purchase Price'] = Purchase_df["Average Purchase Price"].map("${:,.2f}".format)
Purchase_df['Total Purchase Value'] = Purchase_df["Total Purchase Value"].map("${:,.2f}".format)
Purchase_df['Purchase Per Person'] = Purchase_df["Purchase Per Person"].map("${:,.2f}".format)

Purchase_df
```
